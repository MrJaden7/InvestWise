```python

relevant_articles = []

keywords = [
    'revenue', 'earnings', 'profits', 'sales', 'income', 'turnover', 'margins',
    'growth', 'expansion', 'increase', 'surge', 'uptrend', 'rise',
    'loss', 'decline', 'decrease', 'drop', 'setback', 'downturn',
    'investment', 'dividends', 'shareholder returns', 'capital', 'funding', 'acquisition', 'buyback',
    'merger', 'acquisition', 'partnership', 'collaboration', 'layoffs', 'restructuring', 'downsizing',
    'product launch', 'new product', 'innovation', 'recall', 'supply chain', 'market entry',
    'investor confidence', 'market reaction', 'stock performance', 'analyst ratings', 'share price',
    'customer satisfaction', 'public opinion', 'brand reputation', 'customer feedback', 'reviews',
    'market share', 'competitive advantage', 'benchmarking', 'trends', 'industry analysis',
    'competitor performance', 'industry leader', 'new entrants', 'market disruption',
    'economic growth', 'recession', 'inflation', 'market volatility', 'economic outlook',
    'corporate social responsibility', 'sustainability', 'ethical practices', 'governance', 'compliance'
]

for article in response['articles']:
    if any(keyword.lower() in article['title'].lower() or keyword.lower() in article['description'].lower() for keyword in keywords):
        relevant_articles.append({
            'source': article['source']['name'],
            'author': article['author'] if article['author'] else "N/A",
            'title': article['title'],
            'description': article['description'],
            'url': article['url'],
            'urlToImage': article['urlToImage'],
            'publishedAt': article['publishedAt'],
            'content': article['content'] if article['content'] else "N/A"  
        })


if relevant_articles:
    for article in relevant_articles:
        print(f"Title: {article['title']}")
        print(f"Description: {article['description']}")
        print(f"Content: {article['content']}")
        print(f"Link: {article['url']}")
        print(f"Published At: {article['publishedAt']}\n")
else:
    print("No relevant articles found.")
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk

nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()

def vader_sentiment(text):
    sentiment = sia.polarity_scores(text)
    return sentiment['compound']  

df['combined_text'] = df['title'] + ' ' + df['description']


df['vader_sentiment'] = df['combined_text'].apply(vader_sentiment)


df['vader_sentiment_category'] = df['vader_sentiment'].apply(lambda x: 'Positive' if x > 0 else ('Negative' if x < 0 else 'Neutral'))


print(df[['title', 'description', 'vader_sentiment', 'vader_sentiment_category']].head())


average_sentiment = df['vader_sentiment'].mean()


if average_sentiment > 0:
    overall_sentiment = 'Positive'
elif average_sentiment < 0:
    overall_sentiment = 'Negative'
else:
    overall_sentiment = 'Neutral'


print(f"Average Sentiment Score: {average_sentiment}")
print(f"Overall Sentiment Category: {overall_sentiment}")
```
